### import必要套件

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets , preprocessing , linear_model
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
```

讀取訓練資料檔案(csv)

```python
train_data=pd.read_csv("train.csv")   # 將cvs檔案以pandas開啟
train_data
```

```python
train_data['Age']=train_data['Age'].fillna(int(train_data['Age'].mean()))
train_data['Fare']=train_data['Fare'].fillna(int(train_data['Fare'].mean()))

lben=preprocessing.LabelEncoder() #將文字轉換成數字 
train_data['Sex']=lben.fit_transform(train_data['Sex'])
train_data['Embarked']=lben.fit_transform(train_data['Embarked'])

train_data
```

```python
x=pd.DataFrame([train_data['Pclass'],train_data['Sex'],train_data['Age'],train_data['SibSp'],train_data['Parch'],train_data['Fare'],train_data['Embarked']]).T
y=train_data['Survived']
model=XGBClassifier()
model.fit(x,y)
print('XGB訓練準確度',model.score(x,y))
RFmodel = RandomForestClassifier()
RFmodel.fit(x,y)
print('RandomForest訓練準確度',RFmodel.score(x,y))
DTmodel = DecisionTreeClassifier()
DTmodel.fit(x,y)
print('DecisionTree訓練準確度',DTmodel.score(x,y))
```

```python
test_df=pd.read_csv('test.csv')
test_df['Age']=test_df['Age'].fillna(int(train_data['Age'].mean()))
test_df['Sex']=lben.fit_transform(test_df['Sex'])
test_df['Embarked']=lben.fit_transform(test_df['Embarked'])
test_df['Fare']=test_df['Fare'].fillna(int(test_df['Fare'].mean()))
x_test=pd.DataFrame([test_df['Pclass'],test_df['Sex'],test_df['Age'],test_df['SibSp'],test_df['Parch'],test_df['Fare'],test_df['Embarked']]).T
y_test=test_df['Survived']
print('XGB測試準確度',model.score(x_test,y_test))
print('RandomForest測試準確度',RFmodel.score(x_test,y_test))
print('DecisionTree訓練準確度',DTmodel.score(x_test,y_test))
```
